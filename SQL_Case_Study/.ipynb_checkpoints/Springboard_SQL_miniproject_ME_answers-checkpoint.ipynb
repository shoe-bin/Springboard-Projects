```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
f_df = pd.read_sql_query('SELECT * FROM Facilities', engine)
```

```python
m_df = pd.read_sql_query('SELECT * FROM Members', engine)
```

```python
b_df = pd.read_sql_query('SELECT * FROM Bookings', engine)
```

```python
merged_df = b_df.merge(m_df, how='left', on='memid').merge(f_df, how= 'left', on='facid')
```
